```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL yang akan digunakan
base_url = "https://openlibrary.org/trending/forever?page="

# Buat array kosong untuk menyimpan data
titles = []
authors = []
publish_years = []

# Loop dari halaman 1 sampai 3
for page in range(1, 4):  
    url = f"{base_url}{page}"   
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    details = soup.find_all('div', class_='details')
    
    # Loop melalui setiap elemen 'details' untuk mengambil informasi
    for detail in details:    
        title = detail.find('h3', class_='booktitle').text.strip()
        titles.append(title)

        author = detail.find('span', class_='bookauthor').text.replace('by ', '').strip()
        authors.append(author)

        publish_year = detail.find('span', class_='publishedYear').text.replace('First published in ', '').strip()
        publish_years.append(publish_year)
```

```python
# Membuat dan menampilkan dataframe menggunakan pandas
data = {
    'Title': titles,
    'Author': authors,
    'Publish Year': publish_years
}
df = pd.DataFrame(data)
df
```
