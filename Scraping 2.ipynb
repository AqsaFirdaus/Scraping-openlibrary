```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
continents_page = requests.get("https://openlibrary.org/works/OL257943W/A_Game_of_Thrones").text
continents_soup = BeautifulSoup(continents_page, 'html.parser')
title_tag = continents_soup.find('h1', {'class': 'work-title', 'itemprop': 'name'})
title = title_tag and title_tag.get_text(strip=True)
title
```

```python
continents_soup.find('a', itemprop='author').text
```

```python
description_tag = continents_soup.find('div', {'class': 'read-more__content'})
paragraphs = description_tag.find_all('p')
description = '\n\n'.join([p.get_text(strip=True) for p in paragraphs])
description
```

```python
date_tag = continents_soup.find('span', {'itemprop': 'datePublished'})
date_published = date_tag.get_text(strip=True)
date_published
```

```python
subjects = [a.text for a in continents_soup.find_all('a', {'data-ol-link-track': 'BookOverview|SubjectClick'})]
subjects
```

```python
reviews = continents_soup.find_all('span', class_='review__category')
review_data = []

for review in reviews:
    label = review.find('span', class_='reviews__label').text.strip()

    for pill in review.find_all('span', class_='reviews__pill'):
        value = pill.find('span', class_='reviews__value').text.strip()
        percentage = pill.find('span', class_='percentage').text.strip()
        review_data.append({'Category': label, 'Value': value, 'Percentage': percentage})

review_df = pd.DataFrame(review_data)
output = review_df
output
```

```python
table = continents_soup.find('table', class_='history')

# Mengambil semua baris dalam tabel
rows = table.find_all('tr')

# Mengambil data dari setiap baris
history_data = []
for row in rows:
    date = row.find('td', {'class': 'timestamp'}).get_text(strip=True)
    editor = row.find_all('td', {'class': 'timestamp'})[1].get_text(strip=True)
    detail = row.find('td', {'class': 'detail'}).get_text(strip=True)

    # Menambahkan data ke list history_data
    history_data.append({'Date': date, 'Editor': editor, 'Detail': detail})

# Mengubah list menjadi DataFrame menggunakan pandas
history_df = pd.DataFrame(history_data)

# Menyimpan DataFrame ke dalam variabel untuk penggunaan lebih lanjut
output = history_df

# Jika Anda ingin melihat output di Jupyter Notebook, Anda bisa gunakan
output
```
